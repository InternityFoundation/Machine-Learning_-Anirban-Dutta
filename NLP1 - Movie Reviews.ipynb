```python
documents = []
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        documents.append((movie_reviews.words(fileid),category))
documents[0:5]
```

```python
import random
random.shuffle(documents)
documents[0:5]
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    
    if tag.startswith('J'):
        return wordnet.ADJ
    if tag.startswith('V'):
        return wordnet.VERB
    if tag.startswith('N'):
        return wordnet.NOUN
    if tag.startswith('R'):
        return wordnet.ADJ
    else :
        return wordnet.NOUN
```

```python
from nltk import pos_tag
w = "better"
print(pos_tag(w)) # pos tag needs an array not a string type.
print(pos_tag([w]))
```

```python
from nltk.corpus import stopwords
import string
stops = set(stopwords.words('english'))
punctuations = list(string.punctuation)
stops.update(punctuations)
stops,string.punctuation
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

def clean_reviews(words):
    output_words=[]
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])
            clean_words = lemmatizer.lemmatize(w , pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_words.lower())
    return output_words

            
```

```python
docs = [(clean_reviews(document),category) for document , category in documents]
```

```python
docs[0]
```

```python
training_docs = docs[0:1500]
testing_docs = docs[1500:]
```

```python
all_words = []
for doc in training_docs :
    all_words += doc[0]
```

```python
import nltk
freq = nltk.FreqDist(all_words)
common = freq.most_common(3000)
features = [i[0] for i in common]
print(features)
```

```python
def get_features_dict(words):
    current_feature = {}
    words_set = set(words)
    for w in features :
        current_feature[w] = w in words_set
    return current_feature
```

```python
output = get_features_dict(training_docs[0][0])
output
```

```python
training_data = [(get_features_dict(doc),category)for doc , category in training_docs]
training_data = [(get_features_dict(doc),category)for doc , category in testing_docs]
```

```python
training_data[0]
```

```python
from nltk import NaiveBayesClassifier
classifier = NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(classifier,testing_data)
```

```python
classifier.show_most_informative_features(30)
```
